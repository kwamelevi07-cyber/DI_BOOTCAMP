1. Données structurées
  Les rapports financiers d'une entreprise dans un fichier Excel : Les données y sont organisées de manière ordonnée en lignes et en colonnes, avec des types bien définis (chiffres, dates, textes) facilitant le calcul

   Données d'inventaire dans une base de données relationnelle : Il s'agit de l'exemple type de données structurées. Elles suivent un schéma strict (tables SQL) avec des relations bien définies entre les éléments (ex: ID produit, quantité)

2.Données non structurées
 Photographies téléchargées sur une plateforme de médias sociaux : Les fichiers d'images (JPEG, PNG) sont des médias riches qui ne possèdent pas de format textuel ou relationnel sous-jacent.

Une collection d'articles de presse sur un site web : Même si le site web a une mise en page ordonnée, le cœur de la donnée est du texte libre (langage naturel), qui est par nature non structuré.

Enregistrements d'entretiens issus d'une étude de marché : Qu'il s'agisse de fichiers audio, vidéo ou de transcriptions textuelles mot à mot, le contenu varie librement d'un entretien à un autre sans format fixe

                                        #Exercice 7 : Exercice de transformation

1.Articles de blog sur les expériences de voyageMéthode : Le Traitement du Langage Naturel (NLP) couplé à la Reconnaissance d'Entités Nommées (NER). On utilise un Large Language Model (LLM) pour extraire des champs précis et les enregistrer dans un fichier au format JSON ou CSV.Champs structurés obtenus : Nom de la destination, durée du séjour, budget estimé, activités mentionnées, note de sentiment (positif/négatif).Justification : Le texte des blogs est rédigé en langage naturel varié. Le NLP est l'unique approche capable de comprendre le contexte linguistique afin d'isoler des entités géographiques ou sémantiques précises sans nécessiter un format d'écriture uniforme.

 2.Enregistrements audio des appels clientMéthode : Une pipeline combinant la Transcription Speech-to-Text (STT) suivie d'une Analyse de Sentiment automatique. Le logiciel convertit la voix en texte brut, puis un algorithme catégorise l'appel.Champs structurés obtenus : ID de l'appelant, motif de l'appel (technique, facturation, résiliation), durée de l'appel, score de satisfaction client, mots-clés récurrents.Justification : Les fichiers audio ne sont pas exploitables directement par des bases de données relationnelles. La transcription textuelle intermédiaire est indispensable avant d'appliquer des filtres de classification quantitatifs.

3.Notes manuscrites de brainstormingMéthode : La Reconnaissance Optique de Caractères (OCR) intelligente (ou ICR, spécialisée dans l'écriture manuscrite). Les blocs de texte numérisés sont ensuite soumis à un algorithme de clustering (regroupement de données) pour catégoriser les idées.Champs structurés obtenus : Date de la session, nom du contributeur, catégorie de l'idée (marketing, produit, RH), liste des actions à mener.Justification : L'OCR transforme les pixels d'une image ou d'un scan en chaînes de caractères lisibles par une machine. Sans cette conversion de base, le texte reste prisonnier du format image et est impossible à indexer.

4.Tutoriel vidéo sur la cuisineMéthode : L'Analyse Multimodale par IA. Elle combine la vision par ordinateur (pour reconnaître les actions et les ingrédients à l'écran), l'extraction des métadonnées vidéo et le traitement de la piste audio (transcription de la voix du chef).Champs structurés obtenus : Titre de la recette, liste des ingrédients, quantités détectées, temps de cuisson total, étapes séquentielles (étape 1, étape 2, etc.).Justification : Une vidéo superpose des flux visuels et auditifs. L'analyse multimodale permet de synchroniser ce que le chef dit (piste audio) avec ce qu'il fait (flux vidéo), garantissant une extraction fiable de la chronologie de la recette.
